### Entropia e informacion

Para facilitar el proceso, se realizaron funciones con las cuales se realizan las iteraciones y el calculo del indice Gini asi como su ganancia para encontrar el mejor atributo de division, calculados por la ecuaciones:

$$
 Gini(y,s) = 1- \sum_{cj \in dom(y)} \left( \frac{| \sigma _{y=c_{j}}(S)|}{|S|} \right)
$$

$$
GiniGain(a_i, S) = Gini(y, S) - \sum _{v_{i, j} \in dom(a_i)} \frac{|\sigma _{a_i = v_{i, j}}(S)|}{|S|} \cdot Gini(y, \sigma _{a_i, =v_{i,j}}(S))
$$

````python
Atributos =  df.columns

def ite(j):  #Columna objetivo y sus proporciones
    iteraciones = df[Atributos[j]].value_counts(True)
    return iteraciones

def gini(iter): #Indice Gini
    Gini = 1-(sum((it)**2 for it in iter))
    return Gini

print(ite(-1))
print(gini(ite(-1)))

````
